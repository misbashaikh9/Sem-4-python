```python
oldname={'Delhi Daredevils':'Delhi Capitals',
         'Kings XI Punjab':'Punjab Kings',
         'Rising Pune Supergiants':'Pune Warriors',
         'Gujarat Titans':'Gujarat Lions'}
ip=ip.replace(oldname.keys(),oldname.values())
```

```python
ip.Team1.unique()
```

```python
ip.nunique()
```

```python
teams=ip.Team1.unique()
l=[]
for team in teams:
    played=ip[(ip['Team1']==team) | (ip['Team2']==team)].shape[0]
    win=ip[(ip.WinningTeam ==team)].shape[0]
    Home_win=ip[(ip.WinningTeam ==team) & (ip.Team1==team)].shape[0]
    Home_play=ip[ip.Team1==team].shape[0]
    away_win=ip[(ip.WinningTeam ==team) & (ip.Team2==team)].shape[0]
    away_play=ip[ip.Team2==team].shape[0]
    l.append([team,played,((win/played)*100),(Home_win/Home_play)*100,(away_win/away_play)*100])
x=pd.DataFrame(l)
print(x)
```

```python
teams=ip.Team1.unique()
l=[]
for team in teams:
    season=ip[ip['Season']==2022]
    played=ip[((ip['Team1']==team) | (ip['Team2']==team)) & (ip.Season=='2022')].shape[0]
    win=ip[(ip.WinningTeam==team) & (ip.Season=='2022')].shape[0]
    l.append([team,played,win,win*2])
x=pd.DataFrame(l)
print(x)
ip.sort_values('Season')
```
