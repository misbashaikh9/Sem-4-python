```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df.isna().sum()
```

```python
df['Cr']=df.Cr.fillna(df.Cr.median())
```

```python
df['HDL']=df.HDL.fillna(df.HDL.mode()[0])
```

```python
df['AGE']=df.AGE.fillna(df.AGE.mean())
```

```python
df.isna().sum()
```

```python
df=df.dropna()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df.Gender.unique()
```

```python
df.Gender.unique()
```

```python
df['Gender']=df.Gender.replace('f','F')
```

```python
plt.boxplot(df.Cr)
```

```python
Q1=df['Cr'].quantile(0.25)
Q3=df['Cr'].quantile(0.75)
IQR=Q3-Q1
UL=Q3+1.5*IQR
LL=Q1-1.5*IQR
df=df[(df['Cr']>=LL ) & (df['Cr']<=UL)]
```

```python
plt.boxplot(df[['Cr']])
plt.show()
```

```python
#find co relationship in given dataset
df.corr() 
```
