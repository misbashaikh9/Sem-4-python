```python
fig,ax=plt.subplots()
data={'year':[21,22,23,24,25,26],
     'sale':[100,120,150,180,210,240],
     'expense':[70,150,120,210,200,270]}
df=pd.DataFrame(data)
ax.fill_between(df['year'],df['sale'],df['expense'])
ax.plot(df['year'],df['sale'],color='red')
ax.plot(df['year'],df['expense'],color='green')
plt.show()
```

```python
fig,ax=plt.subplots()
data={'year':[21,22,23,24,25,26],
     'sale':[100,120,150,180,210,240],
     'expense':[70,150,120,210,200,270]}
df=pd.DataFrame(data)
ax.fill_between(df['year'],df['sale'],df['expense'],where=df.sale>df.expense,color='black',interpolate=True)
ax.fill_between(df['year'],df['sale'],df['expense'],where=df.sale<df.expense,color='green',interpolate=True)
ax.plot(df['year'],df['sale'],color='red')
ax.plot(df['year'],df['expense'],color='green')
plt.show()
```

```python
data=np.random.randint(low=1,high=100,size=(10,10))
print(data)
```

```python
sns.heatmap(data=data,vmax=60,vmin=40)
plt.show()
```

```python
sns.heatmap(data=data,vmax=60,vmin=40,cmap='tab20',center=0)
plt.show()
```

```python
sns.heatmap(data=data,vmax=60,vmin=40,cmap='tab20',center=0,annot=True)
plt.show()
```

```python
sns.heatmap(data=data,vmax=60,vmin=40,cmap='tab20',center=0,annot=True,linewidths=2,linecolor='black')
plt.show()
```
