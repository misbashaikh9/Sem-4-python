```python
fig,ax=plt.subplots()
data={'year':[21,22,23,24,25,26],
     'sale':[100,120,150,180,210,270],
     'expense':[70,150,120,210,200,270]}
df=pd.DataFrame(data)
ax.stackplot(df['year'],df['sale'],df['expense'],labels=['s','e'])
ax.legend()
plt.show()
```

```python
fig,ax=plt.subplots()
data={'year':[21,22,23,24,25,26],
     'sale':[100,120,150,180,210,240],
     'expense':[70,150,120,210,200,270]}
df=pd.DataFrame(data)
ax.fill_between(df['year'],df['sale'],df['expense'])
plt.show()
```

```python
fig,ax=plt.subplots()
data={'year':[21,22,23,24,25,26],
     'sale':[100,120,150,180,210,240],
     'expense':[70,150,120,210,200,270]}
df=pd.DataFrame(data)
ax.fill_between(df['year'],df['sale'],df['expense'])
ax.plot(df['year'],df['sale'],color='red')
ax.plot(df['year'],df['expense'],color='green')
plt.show()
```

```python
fig,ax=plt.subplots()
data={'year':[21,22,23,24,25,26],
     'sale':[100,120,150,180,210,240],
     'expense':[70,150,120,210,200,270]}
df=pd.DataFrame(data)
ax.fill_between(df['year'],df['sale'],df['expense'],where=df.sale>df.expense,color='black',interpolate=True)
ax.fill_between(df['year'],df['sale'],df['expense'],where=df.sale<df.expense,color='green',interpolate=True)
ax.plot(df['year'],df['sale'],color='red')
ax.plot(df['year'],df['expense'],color='green')
plt.show()
```

```python
stopwords=set(STOPWORDS)
print(stopwords)
stopwords.add('said')
```

```python
url="https://www.gutenberg.org/files/11/11-0.txt"
response=requests.get(url)
text=response.text
print(text)
```

```python
alice_wc=WordCloud(background_color='white',max_words=2000,stopwords=stopwords)
alice_wc.generate(text)
plt.imshow(alice_wc)
plt.axis()
plt.show()
```

```python
# network
G=nx.Graph()
G.add_node('A')
G.add_node('B')
G.add_node('C')
G.add_edge('A','B')
G.add_edge('B','C')
G.add_edge('C','A')
G.add_edge('A','A')
nx.draw(G,node_size=850,node_color='red',width=7,with_labels=True,edge_color='blue')
plt.show()
```

```python
# network
G=nx.DiGraph()
G.add_node('A')
G.add_node('B')
G.add_node('C')
G.add_edge('A','B')
G.add_edge('B','C')
G.add_edge('C','A')
G.add_edge('A','A')
nx.draw(G,node_size=850,node_color='red',width=7,with_labels=True,edge_color='blue')
plt.show()
```
